```python
%matplotlib inline

from matplotlib import pyplot as plt 
import numpy as np

m = 2.6

xs = np.linspace(-10, 10)
zs = m * xs

def mse(w):
    m = len(xs)
    return np.sum([(zs[i] - (w*xs[i]))**2 for i in range(m)])/m

ws = np.linspace(0, 8)
Js = [mse(w) for w in ws]
plt.figure()
plt.xlabel('x')
plt.ylabel('z')
plt.annotate('z = ' + str(m) + 'x', (0, 0), (1, -1))
plt.plot(xs, zs)

plt.figure()
plt.xlabel('w')
plt.ylabel('J(w)')
plt.plot(ws, Js)
line = plt.axvline(m, color='red')
plt.annotate('minimum', (2.6, 500), (2.65, 500), color='darkred')
plt.show()
```
